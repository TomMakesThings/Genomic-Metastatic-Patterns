# Investigation of Mutation Data

As the data is curated from targeted gene sequencing panels, some tumour types are rather sparse in the mutation data. Therefore I plotted the average number of mutations per sample for each tumour type to find which types are most likely to have detectable mutational signatures.

```python
# Open the mutation data
annotated_mutation_df = pd.read_table(directory + "\\Plot_2\\OncoKB\\oncokb_annotated_mutations.txt")
mutation_df = pd.read_table(directory + "\\Data\\data_mutations.txt")
# Add labels to mutation data
mutation_df = mutation_df.join(annotated_mutation_df.loc[:,["Sample_ID", "Subtype", "Sample_Type"]])

mutation_df.head(5)
```

```python
# List of all unique tumour subtypes
all_subtypes = [x for x in np.unique(list(mutation_df["Subtype"])) if x != 'nan']

# Calculate the average number of mutations in primary/metastatic samples for each subtype
subtype_primary_mutations = {}
subtype_metastasis_mutations = {}

for subtype in all_subtypes:
    # Get the rows in sample data for the subtype
    subtype_samples = mutation_df.loc[mutation_df["Subtype"] == subtype]
    
    primary_samples = subtype_samples.loc[subtype_samples["Sample_Type"] == "Primary"]
    metastasis_samples = subtype_samples.loc[subtype_samples["Sample_Type"] == "Metastasis"]
    
    # Count the average number of mutations per sample in the subtype
    subtype_primary_mutations[subtype] = np.mean(list(Counter(list(primary_samples["Sample_ID"])).values()))
    subtype_metastasis_mutations[subtype] = np.mean(list(Counter(list(metastasis_samples["Sample_ID"])).values()))

# Format data in dataframe for plot
mutation_count_df = pd.DataFrame({"Subtype": all_subtypes,
                                  "Primary": list(subtype_primary_mutations.values()),
                                  "Metastasis": list(subtype_metastasis_mutations.values())})
mutation_count_df.head(5)
```

```python
# Plot grouped bar chart of average number of mutations per tumour type
mutation_count_bar = mutation_count_df.plot(x = "Subtype",
                                            kind = "bar",
                                            stacked = False,
                                            title = "Average Number of Mutations per Sample",
                                            figsize = (12, 6))
plt.xlabel("Tumour Type")
plt.ylabel("Number of Mutations")
plt.savefig(directory + '\\COSMIC\\Mutation_Count_Barchart.pdf', bbox_inches = 'tight')
plt.show()
```

# MAF Input File Set up
### Formatting
The matrix generator code below does not run directly on the provided mutation data. Therefore I set the mutation data to be the same format as the example Mutation Annotation Format (MAF) data, downloadable from [SigProfilerMatrixGenerator wiki](https://osf.io/s93d5/wiki/3.%20Using%20the%20Tool%20-%20Input/).

```python
# Open the example data
example_maf_df = pd.read_table(directory + "\\COSMIC\\MAFs\\Example_MAF\\example.maf")
example_maf_df.head(5)
```

```python
# Open the mutation data
mutation_df = pd.read_table(directory + "\\Data\\data_mutations.txt")
mutation_df.head(5)
```

```python
# Subset only columns present in example data
mutation_df_subset = mutation_df.loc[:,["Hugo_Symbol", "Entrez_Gene_Id", "Center", "NCBI_Build",
                                        "Chromosome", "Start_Position", "End_Position", "Strand",
                                        "Variant_Classification", "Variant_Type", "Reference_Allele",
                                        "Tumor_Seq_Allele1", "Tumor_Seq_Allele2", "dbSNP_RS",
                                        "dbSNP_Val_Status", "Tumor_Sample_Barcode"]]
# Rename columns to match
mutation_df_subset.columns = np.array(example_maf_df.columns)[:-1]
mutation_df_subset.head(5)
```

```python
# Save the modified mutation data with all samples as a tab-delimited MAF file
mutation_df_subset.to_csv(directory + "\\COSMIC\\MAFs\\PanCancer\\mutations.maf",
                          sep = "\t", quoting = csv.QUOTE_NONE, index = False)
```

### Splitting MAF File by Subtype
Rather than run on all the whole pan-cancer data, split the mutations by subtype and save each to a new MAF file.

```python
# Open data of tumour samples
samples_data = pd.read_table(directory + "\\Data\\data_clinical_sample.txt", skiprows = 4)
samples_data.head(5)
```

```python
# List of all unique tumour subtypes
all_subtypes = np.unique(samples_data["SUBTYPE"])
# Set directory to store all subtype MAFs
all_subtypes_directory = directory + "\\COSMIC\\MAFs\\Subtypes\\"

for subtype in all_subtypes:
    # Get the rows in sample data for the subtype
    subtype_samples = samples_data.loc[samples_data["SUBTYPE"] == subtype]
    # Find the IDs for all tumour samples belonging to the subtype
    sample_ids = np.unique(subtype_samples["SAMPLE_ID"])
    # Get all mutations belonging to samples of the subtype
    subtype_mutations = mutation_df_subset.loc[mutation_df_subset["Tumor_sample"].isin(sample_ids)]
    # Find the abbreviated name of the subtype
    subtype_abbreviation = list(subtype_samples["SUBTYPE_ABBREVIATION"])[0]
    # Set a directory name to save the mutations
    subtype_directory = all_subtypes_directory + subtype.replace(" ", "_")
    
    # Make a directory to store the subtype mutations
    if not os.path.exists(subtype_directory):
        os.makedirs(subtype_directory)
    
    # Save the subtype's mutations to a new MAF file
    subtype_mutations.to_csv(subtype_directory + "\\" + subtype_abbreviation.replace(" ", "_") + "_MAF.maf",
                             sep = "\t", quoting = csv.QUOTE_NONE, index = False)
```

### MAF File for Uterine Hypermutated
Create separate MAF files for primary and metastatic samples belonging to the uterine hypermutated subtype. This is because this subtype has a large disparity in the number of mutations between primary and metastatic samples.

```python
# Get the rows in sample data for the subtype
uterine_hyper_samples = samples_data.loc[samples_data["SUBTYPE"] == "Uterine Hypermutated"]

uterine_primary = uterine_hyper_samples.loc[uterine_hyper_samples["SAMPLE_TYPE"] == "Primary"]
uterine_metastasis = uterine_hyper_samples.loc[uterine_hyper_samples["SAMPLE_TYPE"] == "Metastasis"]

# Find the IDs for all tumour samples belonging to the subtype
uterine_primary_ids = np.unique(uterine_primary["SAMPLE_ID"])
uterine_metastasis_ids = np.unique(uterine_metastasis["SAMPLE_ID"])

# Get all mutations belonging to samples of the subtype
uterine_primary_mutations = mutation_df_subset.loc[mutation_df_subset["Tumor_sample"].isin(uterine_primary_ids)]
uterine_metastasis_mutations = mutation_df_subset.loc[mutation_df_subset["Tumor_sample"].isin(uterine_metastasis_ids)]

# Find the abbreviated name of the subtype
uterine_hyper_abbreviation = list(uterine_hyper_samples["SUBTYPE_ABBREVIATION"])[0]
# Set a directory name to save the mutations
uterine_hyper_directory = directory + "\\COSMIC\\MAFs\\Primary_Metastatic\\" + "Uterine_Hypermutated"
    
# Make a directory to store the subtype mutations
if not os.path.exists(uterine_hyper_directory):
    os.makedirs(uterine_hyper_directory)
    
# Save the subtype's mutations to a new MAF file
uterine_primary_mutations.to_csv(uterine_hyper_directory + "\\Primary\\" + uterine_hyper_abbreviation.replace(" ", "_") +
                                 "_Primary_MAF.maf", sep = "\t", quoting = csv.QUOTE_NONE, index = False)
uterine_metastasis_mutations.to_csv(uterine_hyper_directory + "\\Metastatic\\" + uterine_hyper_abbreviation.replace(" ", "_") +
                                    "_Metastatic_MAF.maf", sep = "\t", quoting = csv.QUOTE_NONE, index = False)
```

```python
print("Uterine hypermutated has " + str(uterine_primary_mutations.shape[0]) + " mutations in primary samples" +
      " and " + str(uterine_metastasis_mutations.shape[0]) + " in metastatic samples")
```

# Signature Profiler Matrix Generator
### Reference Genome
The mutations data was generated using reference genome GRCh37. Therefore, GRCh37 was downloaded to be used when generating the mutational matricies required for signature analysis.

```python
# Whether to download the reference
download_genome = False

if download_genome:
    # Use FTP to download the reference genome
    wget.download("ftp://alexandrovlab-ftp.ucsd.edu/pub/tools/SigProfilerMatrixGenerator/GRCh37.tar.gz", 
                  directory + "\\COSMIC\\GRCh37.tar.gz")
    
# Set the reference genome from the locally downloaded file
genInstall.install("GRCh37", offline_files_path = directory + "\\COSMIC\\")
```

### Subtype Matricies
Run the COSMIC Signature Profiler Matrix Generator to create mutational matrices for somatic mutations per subtype.

```python
# Find all subtype MAF files
for file_path in glob.glob(all_subtypes_directory + '**/*.maf'):
    # Get the file name and directory
    split_path = file_path.split("\\")
    maf_file = split_path[-1].split(".")[0]
    maf_directory = 'C:\\' + '\\'.join(split_path[1:len(split_path) - 1])
    
    # Create mutational matrices for the subtype
    matrices = matGen.SigProfilerMatrixGeneratorFunc(maf_file, "GRCh37", maf_directory + "\\",
                                                     plot = True, exome = False, bed_file = None,
                                                     chrom_based = False, tsb_stat = False,
                                                     seqInfo = False, cushion = 100)
```
